```python
# 2번
s = input("문자열을 입력하시오 :")
big = ""
small = ""
for i in s:
  if i >= 'A' and i <= 'Z':
    big += i
  else: small += i
print('수정된 문자열 :', (small + big))
```

```python
# 3번
s = input("문자열을 입력하시오 :")
big = 0
small = 0
num = 0
etc = 0
for i in s:
  if i >= 'A' and i <= 'Z': big += 1
  elif i >= 'a' and i<= 'z': small += 1
  elif i >= '0' and i<= '9': num += 1
  else: etc += 1
print("대문자 = %d\n소문자 = %d\n숫자 = %d\n특수문자 = %d\n" %(big, small, num, etc))
```

```python
# 4 번
a = 'ABCD'
b = '1234'
new_str1 = ''
new_str2 = ''

for i in range(len(a)):
  new_str1 += a[i] + b[i]
  new_str2 += a[i] + b[3 - i]
print('new_str1 = %s\nnew_str2 = %s' % (new_str1, new_str2))
```

```python
# 6 번
import re

s = re.findall(r'\d+', 'English = 89, Science = 90, Math = 92, History = 80.')
sum = 0
for i in s:
  sum += int(i)
print('총점 : %d\n평균점수 : %d' %(sum, sum / len(s)))
```

```python
# 7 번
s = input('문장을 입력하시오 :')
key = int(input('이동시킬 칸 수를 입력하시오 :'))

res = ''
for i in s:
  if i >= 'A' and i <= 'Z' or i >= 'a' and i <= 'z':
    res += str(chr(ord(i) + 3))
  else: res += i

print('암호화된 문장 :', res)
```

```python
# 8 번
s = re.findall(r'[0-9a-zA-Z]+', input('문장을 입력하시오 :'))
num = ''
eng = ''
etc = ''
for i in s:
  if i.isdigit(): num += i + ' '
  elif i.isalpha(): eng += i + ' '
  else: etc += i + ' '
print("영문 단어 : %s\n숫자 : %s\n영문자+숫자 : %s\n" % (eng, num, etc))
```

```python
# 9 번
import wikipedia
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

wiki = wikipedia.page('Korea')
text = wiki.content
stop_words = STOPWORDS.union({'south','north','korean','world','country'})

wordcloud = WordCloud(width = 2000, height = 1500, stopwords = stop_words).generate(text)

plt.figure(figsize=(40,30))
plt.imshow(wordcloud)
plt.show()
```
